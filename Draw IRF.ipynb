##### __Draw IRF__

###### Generate time array

```python
horizon = np.arange(0, df.shape[0])
horizon
```

###### Test
Response of FFR to MP shock

```python
plt.figure(figsize=(8, 8))
#  Draw the horizontal line y=0
plt.axhline(0, color="#FF6464",linestyle='dashed', lw=1)

#  Draw median value of IRF
plt.plot(horizon, 
         df['TOTAL_RESULT(1)'],
         color="#17486E", 
         lw=4.5,
         label='Median')

#  Draw upper and lower bound of IRF
plt.fill_between(horizon,
                 df['TOTAL_RESULT(2)'],
                 df['TOTAL_RESULT(3)'],
                 alpha=0.15, color="royalblue",
                label='90% CI')

plt.legend(loc='upper right', fontsize=18)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```

###### Define a function to draw each IRF subgraph

```python
def sub_IRF(x_,legend=False):
    
    #  Draw the horizontal line y=0
    plt.axhline(0, color="#FF6464",linestyle='dashed', lw=1.6)
    
    #  Draw median value of IRF
    plt.plot(horizon, 
         df['TOTAL_RESULT({})'.format(3*(x_-1)+1)],
         color="#17486E",
         lw=4.5,
         label='Median')
    
    #  Draw median value of IRF
    plt.fill_between(horizon,
                     df['TOTAL_RESULT({})'.format(3*(x_-1)+2)],
                     df['TOTAL_RESULT({})'.format(3*(x_-1)+3)],
                     alpha=0.2, color="royalblue",
                     label='90% CI')
    if legend:
        plt.legend(loc='upper right', fontsize=26)
        
    plt.title(names_ls[x_ - 1], fontsize=20)
    plt.yticks(fontsize=15)
    plt.xticks(fontsize=15)
```

###### Draw and save figure
IRF of FFR and other 19 Macroeconomic variables in __*xdata.xlsx*__ to MP schock

```python
plt.figure(figsize=(30, 24))
for i_ in range(1, len(names_ls) + 1):
    plt.subplot(4,5,i_)
    if i_ != 1:
        sub_IRF(i_)
    else:
        sub_IRF(i_, legend=True)
        
plt.savefig('Impluse response of FFR and other 19 Macroeconomic variables to MP schock.svg',
           dpi=900)
plt.show()
```
